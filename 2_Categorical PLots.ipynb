## Categorical Plots

```python
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

This plot shows mean of the total bill for sex category

As we see average total bill for male is higher than for female

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

Countplot count the occurrences

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

The points outside of the wisker are outliers

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True) # jitter ile her kategorinin yogunlugunu goruruz
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='white')
```
